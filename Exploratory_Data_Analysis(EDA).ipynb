```python
d = {"no": 0, "yes": 1}
df["y"] = df["y"].map(d)
df.head()
```

```python
df[df["y"] == 1].select_dtypes(include='number').mean()
```

```python
acd = round(df[df["y"] == 1]["duration"].mean(), 2)
acd_in_min = acd // 60
print("Average call duration for attracted clients =", acd_in_min, "min", int(acd) % 60, "sec")
```

#Visualization

```python
pd.plotting.scatter_matrix(
    df[["age", "duration", "campaign"]],
    figsize = (15, 15),
    diagonal = "kde")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df["age"], df["campaign"]) # Assuming you want to plot 'age' against 'duration'
plt.xlabel("Age")
plt.ylabel("Campaign")
plt.title("Scatter Plot of Age vs Campaign")
plt.show()
```
